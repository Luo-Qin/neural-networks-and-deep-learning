```python
%matplotlib inline
import numpy as np
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
from tensorflow.python.keras.models import Sequential
from tensorflow.python.keras.layers import Dense
import math
```

```python
# Prepare the training data
x_train = 2*math.pi*np.arange(0,4,0.001)
lenx = len(x_train)
y_train = np.sin(x_train)
```

```python
# Model structure
model = Sequential()
model.add(Dense(12, input_dim=1, kernel_initializer='normal', activation='relu'))
model.add(Dense(12, activation='relu'))
model.add(Dense(12, activation='relu'))
model.add(Dense(1, activation='linear'))
model.summary()
```

```python
# Model compile
model.compile(loss='mse', optimizer='adam', metrics=['mse','mae'])
history = model.fit(x_train, y_train, epochs=150, batch_size=32,  validation_split=0.2)
```

# Test the model

```python
x_test = 2*math.pi*np.arange(0,2,0.01)
y_test = np.sin(x_test)
prediction = model.predict(x_test)
```

```python
plt.rcParams["figure.figsize"] = (6, 4)
plt.style.use("bmh")
fig, ax = plt.subplots(facecolor='w')
plt.plot(x_test,y_test,"o")
plt.plot(x_test,prediction,"s")
```
